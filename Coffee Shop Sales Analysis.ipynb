```python
y = df.groupby([df['transaction_date'].dt.to_period('M'),'store_location']).size().reset_index(name='count')

# Convert Period to Timestamp for plotting
y['transaction_date'] = y['transaction_date'].dt.to_timestamp()

plt.figure(figsize=(20, 6))
for category in y['store_location'].unique():
    subset = y[y['store_location'] == category]
    plt.plot(subset['transaction_date'], subset['count'], label=category, marker='o')

plt.xlabel("Transaction Date")
plt.ylabel("Count")
plt.title("Store Transaction Trends")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for i in df.product_category.unique():
    category = df[df.product_category == i]
    hourly_analysis = category.groupby([category.transaction_date.dt.date,'Hour'])['sales'].sum().reset_index()
    avg_hourly_analysis = hourly_analysis.groupby('Hour')['sales'].mean()
    plt.plot(avg_hourly_analysis.index, avg_hourly_analysis.values,label=i)
    
plt.legend()
plt.show()
plt.savefig('category_hour')
```
